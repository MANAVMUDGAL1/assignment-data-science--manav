```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df
```

reading score and writing score make positive relation with each other

```python
sns.relplot(x='writing score',y='reading score',data=df,hue='race/ethnicity')
plt.title('POSITIVE RELATION')
plt.show()
```

male have given more exam in comparison to female


```python
sns.relplot(x='writing score',y='reading score',data=df,hue='race/ethnicity',col='gender')

plt.show()
```

all group have positive relation but student of group c have given more exam

```python
sns.relplot(x='writing score',y='reading score',data=df,hue='gender',style='race/ethnicity',row='race/ethnicity')
```

all groups have positive relation

```python
sns.relplot(x='writing score',y='reading score',data=df,hue='gender',style='race/ethnicity',col='race/ethnicity',kind='line')
```

group A has most associate degree when test preparation course is none group D has most associate degree when test preparation course is completed

```python
sns.relplot(x='race/ethnicity',y='parental level of education',data=df,hue='gender',style='lunch',col='test preparation course',kind='line')
```
